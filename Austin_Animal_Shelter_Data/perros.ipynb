```python
#function to extract latitude from address list
def coordinates_lat(add_lst):
    lat_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lat_lst.append("nan")
        else:
            lat_lst.append(loc.latitude)
    return lat_lst
```

```python
#function to extract longitude from address list
def coordinates_lon(add_lst):
    lon_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lon_lst.append("nan")
        else:
            lon_lst.append(loc.longitude)
    return lon_lst
```

```python
#transforming location to a list to be used in functions
loc_lst = list(perro["loc"][:100])
#list for all latitudes
lat = coordinates_lat(loc_lst)
```

```python
#list for all longitudes
#lon = coordinates_lon(loc_lst)
loc_lst = list(perro["loc"][:100])
#list for all latitudes
lon = coordinates_lon(loc_lst)
```
