#### Extracting latitude and longitude data from df

```python
from geopy.geocoders import GoogleV3
import geopy.distance
import googlemaps
from geopy.geocoders import Nominatim
geolocator = GoogleV3(api_key=api_key)
```

```python
#function to extract latitude from address list
def coordinates_lat(add_lst):
    lat_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lat_lst.append("nan")
        else:
            lat_lst.append(loc.latitude)
    return lat_lst
```

```python
#function to extract longitude from address list
def coordinates_lng(add_lst):
    lng_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lng_lst.append("nan")
        else:
            lng_lst.append(loc.longitude)
    return lng_lst
```

```python
#transforming location to a list to be used in functions
loc_lst = list(perrito["loc"])
#list for all latitudes
lat = coordinates_lat(loc_lst)
```

```python
#list for all longitudes

loc_lst = list(perrito["loc"])
#list for all latitudes
lng = coordinates_lng(loc_lst)
```

```python
perrito["lat"] = lat
perrito["lng"] = lng
```

```python
def plot_data(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Austin Animal Shelter Intakes', 
             width=bokeh_width, height=bokeh_height)
    # definition of the column data source: 
    source = ColumnDataSource(perrito)
   
    center = p.circle('lng', 'lat', size=5, alpha=1, 
                      color='yellow', source=source)
    show(p)
    return p

p = plot_data(lat_main, lon_main, map_type='terrain')
```
