#### Extracting latitude and longitude data from df

```python
from geopy.geocoders import GoogleV3
import geopy.distance
import googlemaps
from geopy.geocoders import Nominatim
geolocator = GoogleV3(api_key=api_key)
```

```python
#function to extract latitude from address list
def coordinates_lat(add_lst):
    lat_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lat_lst.append("nan")
        else:
            lat_lst.append(loc.latitude)
    return lat_lst
```

```python
#function to extract longitude from address list
def coordinates_lng(add_lst):
    lng_lst = []
    for address in add_lst:
        loc = geolocator.geocode(address)
        if not loc:
            lng_lst.append("nan")
        else:
            lng_lst.append(loc.longitude)
    return lng_lst
```

```python
#transforming location to a list to be used in functions
loc_lst = list(perrito["loc"])
#list for all latitudes
lat = coordinates_lat(loc_lst)
```

```python
#list for all longitudes

loc_lst = list(perrito["loc"])
#list for all latitudes
lng = coordinates_lng(loc_lst)
```

```python
#in order to save time for future testing, the longitude and latitude values were stored in a text file
#function to clean longitude and latitude values and convert to lists without having to extract data using geocode 
def txt_to_list(file):
    with open(file) as f:
        lines = f.readlines()
        values = []
        for line in lines: 
            if line != " 'nan',\n":
                clean_line = float(line.strip(",\n"))
                values.append(clean_line)
            else:
                clean_line = "nan"
                values.append(clean_line)
    return(values)
```

```python
lat_values = txt_to_list("lat.txt")
lng_values = txt_to_list("lng.txt")
```

#### Plotting map with the data from the df

```python
from bokeh.models import ColumnDataSource
```

```python
def plot_data(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Austin Animal Shelter Intakes', 
             width=bokeh_width, height=bokeh_height)
    # definition of the column data source: 
    source = ColumnDataSource(perrito)
   
    center = p.circle('lng', 'lat', size=5, alpha=1, 
                      color='yellow', source=source)
    show(p)
    return p

p = plot_data(lat_main, lon_main, map_type='terrain')
```

```python
def plot_animals(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Austin Animal Shelter Intakes', 
             width=bokeh_width, height=bokeh_height)
    
    source_dogs = ColumnDataSource(dogs)
    center_dogs = p.circle('lng', 'lat', size=5, alpha=1, 
                   source=source_dogs,
          color="green")
    source_cats = ColumnDataSource(cats)
    center_cats = p.triangle('lng', 'lat', size=5, alpha=1, 
                   source=source_cats,
          color="blue")
    source_birds = ColumnDataSource(birds)
    center_birds = p.hex('lng', 'lat', size=5, alpha=1, 
                   source=source_birds,
          color="yellow")
    source_others = ColumnDataSource(others)
    center_others = p.diamond('lng', 'lat', size=5, alpha=1, 
                   source=source_others,
          color="red")
    show(p)
    return p

p3 = plot_animals(lat_main, lon_main, map_type='terrain')
```

#### Plotting map with categorical values from "Intake Category"

```python
from bokeh.models import CategoricalColorMapper
from bokeh.plotting import figure
from bokeh.palettes import Bokeh8
from bokeh.models import Legend
```

```python
def plot_animal_types(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Austin Animal Shelter Intakes', 
             width=bokeh_width, height=bokeh_height)
    mapper = CategoricalColorMapper(factors=list(perrito["Intake Type"].unique()), palette=[Bokeh8[0], Bokeh8[2], Bokeh8[5], Bokeh8[7]])
    source_dogs = ColumnDataSource(dogs)
    center_dogs = p.circle('lng', 'lat', size=5, alpha=1, 
                   source=source_dogs, color={'field': "Intake Type", 'transform': mapper})
    source_cats = ColumnDataSource(cats)
    center_cats = p.triangle('lng', 'lat', size=5, alpha=1, 
                   source=source_cats, color={'field': "Intake Type", 'transform': mapper} )
    source_birds = ColumnDataSource(birds)
    center_birds = p.hex('lng', 'lat', size=5, alpha=1, 
                   source=source_birds, color={'field': "Intake Type", 'transform': mapper})
    source_others = ColumnDataSource(others)
    center_others = p.diamond('lng', 'lat', size=5, alpha=1, 
                   source=source_others, color={'field': "Intake Type", 'transform': mapper})
    ##legend items
    #animal type
    dog_legend = p.circle(size=4, alpha=1, color = "black")
    cat_legend = p.triangle(size=4, alpha=1, color = "black")
    bird_legend = p.hex(size=6, alpha=1, color = "black")
    other_legend = p.diamond(size=6, alpha=1, color = "black")
    #intake type
    intake = p.square(size=6, alpha=1, source=perrito, color={'field': "Intake Type", 'transform': mapper})
    legend_animal_types = Legend(items = [("Dogs", [dog_legend]), ("Cats", [cat_legend]), ("Birds", [bird_legend]), ("Other", [other_legend])])
    legend_intake_type = Legend(items = [(perrito["Intake Type"].unique()[0], [intake]), (perrito["Intake Type"].unique()[1], [intake]), (perrito["Intake Type"].unique()[2], [intake]), (perrito["Intake Type"].unique()[3], [intake])])
    #legend = Legend(items = [(perrito["Intake Type"].unique()[0], [square])], location=(0, -30))
    #legend = Legend(items = [(perrito["Intake Type"].unique()[0], p.square(0, 0, fill_color=Bokeh8[0]))], location=(0, -30))
    #legend = Legend(items=[(perrito["Intake Type"].unique()[0], p.square(0, 0, fill_color=Bokeh8[0]),location=(0, -30))])
#(perrito["Intake Type"].unique()[1], [ Bokeh8[2]]),location=(0, -30)),]
#     (perrito["Intake Type"].unique()[2], [Bokeh8[5]]),], )
    p.add_layout(legend_animal_types, 'right')
    p.add_layout(legend_intake_type, 'below')
    show(p)
    return p

p4 = plot_animal_types(lat_main, lon_main, map_type='terrain')
```
